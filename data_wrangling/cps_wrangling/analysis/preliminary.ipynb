```python
df = sub.minor_xs('flow')
x = df.reset_index()
df = (x.loc[~(x['PULINENO'] == -1)]).set_index(['HRHHID', 'HRHHID2', 'PULINENO'])
print(df.head())
df = df.dropna(how='all', axis=[0, 1])
all_clean = df.dropna(how='any', axis=0)
all_clean.head()
```

```python
# status rather than flow
df = wp.minor_xs('labor_status')
x = df.reset_index()
df = (x.loc[~(x['PULINENO'] == -1)]).set_index(['HRHHID', 'HRHHID2', 'PULINENO'])
print(df.head())
df = df.dropna(how='all', axis=[0, 1])
clean = df.dropna(how='any', axis=0)
clean.head()
```

```python
import vincent
vincent.core.initialize_notebook()
```

```python
labels = {1: 'employed', 2: 'employed', 3: 'unemployed', 4: 'unemployed', 5: 'nilf', 6: 'nilf', 7: 'nilf'}
cts = clean.replace(labels).apply(lambda x: pd.value_counts(x))
cts = cts.sort_index()
```

```python
stacked = vincent.StackedArea(cts.T)
stacked.axis_titles(x='Index', y='Value')
stacked.legend(title='Employment Status')
stacked.display()
```

```python
# With Attrition
labels = {1: 'employed', 2: 'employed', 3: 'unemployed', 4: 'unemployed', 5: 'nilf', 6: 'nilf', 7: 'nilf'}
cts = df.replace(labels).apply(lambda x: pd.value_counts(x))
cts = cts.sort_index()
cts.T

stacked = vincent.StackedArea(cts.T)
stacked.axis_titles(x='Index', y='Value')
stacked.legend(title='Employment Status')
stacked.display()
```
