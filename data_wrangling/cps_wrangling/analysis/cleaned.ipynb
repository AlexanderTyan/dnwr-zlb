```python
# Setup
df['either_history'] = df.either_history.replace(-1, np.nan)

s_dummies = pd.get_dummies(df.quarter).rename(columns={1: "Q1", 2: "Q2",
                                                       3: "Q3", 4: "Q4"})
df = pd.concat([df, s_dummies], axis=1)

# Add time trend
df['trend'] = df.year - 1996 + df.quarter

# Filter out outer .005% of hours and .003% of real hourly earnings
low_hours, high_hours = df.hours.quantile(.005), df.hours.quantile(.995)
low_earn, high_earn = df.real_hr_earns.quantile(.003), df.real_hr_earns.quantile(.997)

trim = df.query('hours > low_hours & hours < high_hours &'
                'real_hr_earns > low_earn & real_hr_earns < high_earn')

# Deseasonalize real hourly wages

seasonal_mod = sm.OLS.from_formula("np.log(real_hr_earns) ~ Q1 + Q2 + Q3", trim)
seasonal_res = seasonal_mod.fit()

seasonal_effects = seasonal_res.params
seasonal_effects['Q4'] = 0

# BUG https://github.com/pydata/pandas/issues/5653
# real_hr_earns_d = df.set_index('quarter', append=True)['real_hr_earns'].copy()
# real_hr_earns_d = real_hr_earns_d.rename(index={1: "Q1", 2: "Q2", 3: "Q3",
#                                                 4: "Q4"}, level='quarter')

real_hr_earns_d = trim[['quarter', 'real_hr_earns']].copy()
real_hr_earns_d['quarter'] = real_hr_earns_d['quarter'].replace({1: "Q1", 2: "Q2",
                                                                 3: "Q3", 4: "Q4"})
real_hr_earns_d = real_hr_earns_d.set_index('quarter', append=True)
# Series for sub
real_hr_earns_d = np.log(real_hr_earns_d['real_hr_earns']).sub(seasonal_effects,
                                                               level='quarter')
idx = real_hr_earns_d.index.droplevel('quarter')  # to match with trim
real_hr_earns_d.index = idx
trim['ln_real_hr_earns_d'] = real_hr_earns_d

# First Stage

trim = add_demo_dummies(trim)
trim['expr_2'] = trim.expr ** 2
trim['expr_3'] = trim.expr ** 3
trim['expr_4'] = trim.expr ** 4
```

### First Stage:

\begin{equation}
    \ln w_{it} = \boldsymbol{x_{i}}^T \boldsymbol{\beta} + \ln \hat{w}_{it}
\end{equation}

Regressing each individual $i$'s wage at time $t$ on demographic variables:

- `age`
- `sex_d`: (female=1)
- `race_d`: (black=1)
- `married_d`: (with spouse present=1)
- `edu_bin`:
        0. Didn't finish highschool
        1. Highschool or GED
        2. Some college
        3. Associate
        4. Finished college
        5. Some graduate
- $expr + expr^2 + expr^3 + expr^4$ experience
- `Q1 .. Q3`: quarterly dummies


```python
mod_fs = sm.OLS.from_formula("ln_real_hr_earns_d ~ age + sex_d + race_d +"
                             "married_d + edu_bin + expr + expr_2 +"
                             "expr_3 + expr_4 + trend", trim.dropna())
res_fs = mod_fs.fit()
res_fs.summary2()
```

### Second Stage

\begin{equation}
    \Delta \ln \hat{w}_{jt} = \alpha_j + \eta_j \Delta \ln y_{jt} + \epsilon_{jt}
\end{equation}

So we are interested in $\eta_j$ the elasticity of the wage with respect to productivity (we're in log differences here).

The $\hat{w}_{jt}$ are the residuals from the first stage, grouped by time and subgroup (existing employed vs. hired from nonemployment). They are aggregated by mean or median.


```python
g = df.reset_index().groupby(['qmonth', 'either_history'])['resid_fs']
mean_resid_fs = g.agg('mean').unstack(['either_history'])[[0, 1]]

prod = df[['productivity']].groupby(level='qmonth').mean()
prod.name = 'productivity'

mean_resid_fs = pd.concat([mean_resid_fs, prod], axis=1)
first_diff = mean_resid_fs.diff().dropna()
first_diff = first_diff.rename(columns={0: 'from_e', 1: 'from_n'})
```

```python
mod_e_ss = sm.OLS.from_formula("np.log(from_e) ~ np.log(productivity)", first_diff)
res_e_ss = mod_e_ss.fit()
res_e_ss.summary()
```

```python
mod_n_ss = sm.OLS.from_formula("np.log(from_n) ~ np.log(productivity)", first_diff)
res_n_ss = mod_n_ss.fit()
res_n_ss.summary()
```

# Via Estimating 2

Above, I estimated the elasticity wrt productivity from using the regressions in the first statge.

Now I'm going to use equation 2:

\begin{equation}
    \ln \hat{w}_{jt} = \ln w_{jt} - \boldsymbol{(x_{jt} - \bar{x}_{j})^T \beta}
\end{equation}

The first part generates the wage index. Break into subgroups $j$ according to history.



```python
g_ts = trim.reset_index().groupby(['qmonth', 'either_history'])
mean_ts = g_ts[res_fs.params.index[1:]].mean()
means = trim.groupby('either_history')[res_fs.params.index[1:]].mean()

deviations = mean_ts.sub(means, level='either_history')

mean_earn = g_ts['ln_real_hr_earns_d'].mean()
```

```python
# employed first
w_et = mean_earn.xs(0, level='either_history')
dev_e = deviations.xs(0, level='either_history')

r_e = dev_e.dot(res_fs.params.iloc[1:])
```

```python
# now unemployed
w_nt = mean_earn.xs(1, level='either_history')
dev_n = deviations.xs(1, level='either_history')

r_n = dev_n.dot(res_fs.params.iloc[1:])
```
