```python
# Regresssion
# TOOD: fixup married
import statsmodels.api as sm
```

```python
model = sm.OLS.from_formula("real_hr_earns ~ age + sex + race + married + productivity", df)


# with seasonal
model = sm.OLS.from_formula("np.log(real_hr_earns) ~ age + sex + race + married + productivity "
                            "+  D_0 + D_1 + D_2", df[df.real_hr_earns > 0])
res = model.fit()
res.summary()
```

```python
# by history
model = sm.OLS.from_formula("real_hr_earns ~ age + sex + race + married + productivity + either_history"
                            "+  D_0 + D_1 + D_2", df[df.real_hr_earns > 0])
res = model.fit()
res.summary()
```
