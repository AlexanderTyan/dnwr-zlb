# Unemployment Volatility, Wage Stickiness, and Worker Flows

## Unemployment Volatility Puzzle (Pissaredes 2009)

- The Puzzle: standard search and matching model can't match volitility of unemployment
- Sticky wages offered as solution (Shimer, Hall & Milgrom)
- Pissaredes: Stickiness of *existing* wages is *irrelevant* for job creation / unemployment
- All that matters is the wage stickiness for *new hires*


## Two Worries

1. Ignores whether entrant is from unemployment vs. nonemployment
2. No on-the-job search


# Current Population Survey

- Monthly survey
- Started in 1940s, redesigned in 1994
- Currently ~ 60,000 individuals broken into 8 waves
- Interviewed for 4 months, out of sample for the next 8, interviewed for 4 more months
- Earnings questions asked in interview months 4 and 8
