# Demographic Time Series

```python
g = df.groupby(level='qmonth')
ts = g[['age', 'sex_d', 'race_d', 'expr', 'edu_bin', 'married_d']].mean().iloc[:-1]
```

```python
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(figsize=(15, 15), nrows=3, ncols=2)
title_fontdict = {'size': 16}

ts['age'].plot(ax=ax1, label='Age')
ax1.set_title('Age (years)', fontdict=title_fontdict)

ts['sex_d'].plot(ax=ax2)
ax2.set_title('Sex (proportion female)', fontdict=title_fontdict)

ts['race_d'].plot(ax=ax3)
ax3.set_title('Race (proprtion nonwhite)', fontdict=title_fontdict)

ts['expr'].plot(ax=ax4)
ax4.set_title('Experience (years)', fontdict=title_fontdict)

ts['edu_bin'].plot(ax=ax5)
ax5.set_title('Education (see appendix)', fontdict=title_fontdict)

ts['married_d'].plot(ax=ax6)
ax6.set_title('Marital Status (proportion married)', fontdict=title_fontdict)

axes = [ax1, ax2, ax3, ax4, ax5, ax6]
for ax in axes:
    ax.set_xlabel('Quarter')
    add_rec_bars(ax)
    plt.setp(ax.get_xticklabels(), fontsize=12)
    plt.setp(ax.get_yticklabels(), fontsize=12)

fig.tight_layout()

plt.savefig('/Users/tom/Economics/inflation_in_recession/dnwr_zlb/writing/cps_rigidity/demographics_ts.pdf', format='pdf')
```
