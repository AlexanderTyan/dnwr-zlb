```python
key = 0.0253571428571, 0.721052631579
ws = wses[key]
g = gs[key]
params['pi'] = key[0], 'a'
params['lambda_'] = key[1], 'b'
z_grid = params['z_grid'][0]
flex_ws = Interp(z_grid, ss_wage_flexible(params, shock=z_grid))
```

```python
# Hack out innards of get_rigid_output
def inside_output(ws, params, flex_ws, g):
    sigma, eta, gamma, pi = (params['sigma'][0],
                             params['eta'][0], params['gamma'][0],
                             params['pi'][0])
    lambda_ = params['lambda_'][0]
    # z_grid = params['z_grid'][0]
    # ln_dist = params['full_ln_dist'][0]
    # shocks = np.sort(shocks)
    dg = kde.gaussian_kde(g.observations.ravel())
    shocks = np.sort(truncated_draw(params, lower=.005, upper=.995,
                                    kind='lognorm', size=1000), axis=0).ravel()

    w_grid = params['w_grid'][0]
    wmax = w_grid[-1]

    p1 = ((1 / shocks) ** (gamma * (eta - 1) / (gamma + eta)) *
          (flex_ws(shocks) / ws(shocks)) ** (eta - 1))

    p2 = ((1 / shocks) ** (gamma * (eta - 1) / (gamma + eta)) *
          g(ws(shocks).ravel() * (1 + pi)) *
          (flex_ws(shocks) / ws(shocks)) ** (eta - 1))

    inner_f = lambda w, z: ((1 + pi) * dg.evaluate(w * (1 + pi))[0] *
                            (flex_ws(z) / w)**(eta - 1))

    w_range = np.sort(ws(shocks))
    sub_w = lambda z: w_range[w_range > ws(z)]  # TODO: check on > vs >=
    p3 = np.zeros(len(shocks))
    for i, z in enumerate(shocks[:-1]):  # empty range for last one
        inner_range = sub_w(z)
        a = inner_range[0]
        inner_vals = quad(inner_f, a, wmax, args=z)[0]
        p3[i] = (1 / z)**(gamma * (eta - 1) / (eta + gamma)) * inner_vals

    p3 = p3

    # z_part is \tilde{Z} in my notes.
    # z_part is decreasing in p1 + p2
    # output is *decreasing* in z_part (it goes in as 1 over)
    # so output is increasing in p1 + p2
    return p1, p2, p3, shocks

def z_part(p1, p2, p3, params):
    lambda_ = params['lambda_'][0]
    eta = params['eta'][0]
    gamma = params['gamma'][0]
    z_t = ((1 - lambda_) * p1 +
            lambda_ * (p2 + p3))**(-(eta + gamma) / (gamma * (eta - 1)))
    return z_t

def output(z_t, params):
    lambda_ = params['lambda_'][0]
    eta = params['eta'][0]
    gamma = params['gamma'][0]
    out = (((eta - 1) / eta)**(gamma / (1 + gamma)) *
           (1 / z_t)**(gamma / (1 + gamma)))
    return out
```

```python
p1, p2, p3, shocks = inside_output(ws, params, flex_ws, g)
```

```python
def make_scatter(p1, p2, p3, shocks):
    colors = it.cycle(plt.rcParams['axes.color_cycle'])
    fig, ax = plt.subplots(figsize=(13,8))
    ax.scatter(shocks, p1, s=8, marker='.', label='p1', color=next(colors))
    ax.scatter(shocks, p2, s=8, marker='.', label='p2', color=next(colors))
    ax.scatter(shocks, p3, s=8, marker='.', label='p3', color=next(colors))
    ax.scatter(shocks, p2 + p3, s=8, marker='.', label='$p2 + p3$', color=next(colors))
    ax.scatter(shocks, output(z_part(p1, p2, p3, params), params), s=8, marker='^', label='output',
               color=next(colors))
    ax.set_title('$\pi={0:.3f}, \lambda={1:.3f}$'.format(*key))
    ax.legend()
    return ax
```
