```python
# just saving for later
cts1, idx1, ax1 = hist(df.diff(1).loc[30], bins='scott', alpha=.35, width=.0005)
t = pd.concat([df.diff(x).iloc[30] for x in [1, 2, 3, 4]], axis=1)
```

### Same $\pi$ and $\lambda$, across periods

```python
fig, ax = plt.subplots(figsize=(13, 8))
cts, idx, other = hist(t.values, histtype='bar', bins=idx1, label=['lag=1', 'lag=2', 'lag=3', 'lag=4'], ax=ax)

pi, lambda_ = [round(x, 2) for x in key]

ax.set_title('Across Periods (normal scale) $\pi={0}$, $\lambda={1}$'.format(pi, lambda_))
plt.legend()

fig, ax = plt.subplots(figsize=(13,8))
cts, idx, other = hist(t.values, histtype='bar', bins=idx1, log=True, label=['lag=1', 'lag=2', 'lag=3', 'lag=4'], ax=ax)


ax.set_title('Across Periods (log scale), $\pi={0}$, $\lambda={1}$'.format(pi, lambda_))
plt.legend()
```

```python
tnorm = pd.DataFrame(cts, columns=idx[1:]).T
tnorm /= tnorm.sum()
# Need to round labels
tnorm.plot(kind='bar', figsize=(13,8), title='As a percentage', rot='45')
```

```python
tlog = pd.concat([np.log(df).diff(x).iloc[30] for x in [1, 2, 3, 4]], axis=1)
```

```python
fig, ax = plt.subplots(figsize=(13,8))
cts, idx, other = hist(tlog.values, histtype='bar', cumulative=True, bins=idx1, label=['lag=1', 'lag=2', 'lag=3', 'lag=4'], ax=ax)
ax.set_title('Cumulative')
ax.set_xlabel('Log Change')
ax.legend(loc=2)
```
