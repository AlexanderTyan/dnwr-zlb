A bit of a different (~~probably better~~) tact:  We can get the steady-state hours works from the wage schedule with a shock $Z = 1$.  That gives us wages which gives us aggregate output which together give us utility.

Actually, I'm down on this approach.  Having rigid wages really sucks when you get a *bad* shock and can't adjust your wage/hours worked.  To that end we should keep giving shocks, getting a path of hours worked, and evalulating utilitity at those hours worked **with the shock that genereated those hours**.  Remember $Z_{it}$ shows up in the utility function.  And we can't just back out $Z_{it}$ since it could be a restricted guy...   I'll get to all that later though.

```python
ls = {key: wses[key](1) for key in wses}  # Each evaluated at Z=1
ls = ar.tuple_constructor(ls)
ls = ls.sort_index()
ls.columns = ['wage']
fig, ax = plt.subplots(figsize=(13, 8))
ax = ls.plot(kind='bar', ax=ax, color=colors, title='Non-Stochastic Wages Grouped by $\pi$')
ax.set_ylim(.99, 1.006)
ax.grid(False)
plt.draw()
```
