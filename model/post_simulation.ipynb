## Fun with densities

```python
from matplotlib.collections import LineCollection
from scipy.stats import kde
```

```python
gps = ar.read_output(all_files, kind='gp')
dgs = [(k, kde.gaussian_kde(gps[k].observations.ravel())) for k in gps]
xs = [dg.dataset.ravel() for k, dg in dgs]
ys = [dg.evaluate(dg.dataset.ravel()) for k, dg in dgs]
```

```python
xys = [np.vstack([x, y]).T for x, y in zip(xs, ys)]
```

```python
colors = plt.cm.gist_heat([lam for pi, lam in gps])
```

```python
col = LineCollection(xys, colors=colors, cmap=plt.cm.gist_heat)
```

```python
fig, ax = plt.subplots(figsize=(13,8))
ax.set_xlim(.925, 1.08)
ax.set_ylim(.35, 60)
ax.add_collection(col)
#plt.colorbar(col.cmap)
```
