```python
def make_hist(pan, subpairs, ax=None, **kwargs):
    t = pd.concat([pan.xs(x, axis='items').diff().iloc[30] for x in subpairs], axis=1)
    t.columns = map(str, subpairs)
    _, idx1, fig = hist(t[t.columns[0]], bins='scott', alpha=.35, width=.0005)
    plt.close()
    if ax is None:
        fig = plt.figure(**kwargs)
        ax = fig.add_subplot(111)
    return hist(t.values, bins=idx, ax=ax, histtype='bar', label=t.columns.tolist())
```

```python
# Across pi, given lambda
fig0 = plt.figure()
make_hist(pan, pairs[0::4], figsize=(13, 8))
plt.legend()

fig1 = plt.figure()
make_hist(pan, pairs[1::4], figsize=(13, 8))
plt.legend()

fig2 = plt.figure()
make_hist(pan, pairs[2::4], figsize=(13, 8))
plt.legend()

fig3 = plt.figure()
make_hist(pan, pairs[3::4], figsize=(13, 8))
plt.legend()
```

```python
def make_panel(wses, params, pairs, log=False):
    dfs = {}
    for key in pairs:
        ws = wses[key]
        pths = sample_path(ws, params, w0=.9, nseries=1000, nperiods=50)
        df = pd.DataFrame(pths)
        if log:
            dfs[key] = np.log(df)
        else:
            dfs[key] = np.log(df)
    pan = pd.Panel(dfs)
    pan.items.name = 'key'
    return pan
```

```python
lpan = make_panel(wses, params, pairs, log=True)
```
