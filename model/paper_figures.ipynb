```python
params = load_params()
all_files = ar.get_all_files(params)
wses = ar.read_output(all_files, kind='ws')
keys = wses.keys()
pis, lambdas = zip(*keys)  # FTW
pis_u, lambdas_u = sorted(set(pis)), sorted(set(lambdas))  # unique
```

```python
# Bring some parameters into the namespace
sigma = params['sigma'][0]
z_grid = params['z_grid'][0]
flex_ws = Interp(z_grid, ss_wage_flexible(params, shock=z_grid))

# Grab the wage schedules to be plotted
key1 = (0.0185714285714, 0.586842105263)
key2 = (0.0525, 0.184210526316)
ws1 = wses[key1]
ws2 = wses[key2]
z_grid = params['z_grid'][0]
flex_ws = Interp(z_grid, ss_wage_flexible(params, shock=z_grid))
```

## Wage Schedule: Comparing Fixed and Rigid

```python
# Include

fig, ax = plt.subplots(figsize=(13, 8))
flex_ws.plot(ax=ax, label='flexible: $\hat{w}_t(Z)$', linewidth=2)
ws2.plot(ax=ax, label='rigid: $w_t(Z),\  \pi={0:.2f}$'.format(key2[0]), linewidth=2)
ws1.plot(ax=ax, label='rigid: $w_t(Z),\  \pi={0:.2f}$'.format(key1[0]), linewidth=2)
ax.set_title('Wage Schedule: Flexible and Rigid', fontsize=18)
ax.text(1.62, .94, '$\lambda={0:.2f}$'.format(key1[1]), fontdict={'fontsize': 16, 'family': 'sans-serif'})
ax.text(1.62, .926, '$\sigma={0:.2f}$'.format(sigma), fontdict={'fontsize': 16, 'family': 'sans-serif'})
ax.grid(False)
ax.legend(loc=2)

ax.set_xlabel('Idiosyncratic shock: $Z_t$', fontdict={'fontsize': 16, 'family': 'sans-serif'})
ax.set_ylabel('Optimal wage choice: $w$', fontdict={'fontsize': 16, 'family': 'sans-serif'})

for item in ax.get_legend().get_texts():
    item.set_fontsize(14)

fig.savefig('./figures/wage_schedule_flex_rigid.pdf', format='pdf')
```
